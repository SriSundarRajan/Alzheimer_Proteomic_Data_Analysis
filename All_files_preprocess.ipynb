```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import seaborn as sns
```

Preprocessing of TMT files. There are totally 8 different file. Each cell represents each single file.

File_TMT1 prprocessing and export preprocessed data

```python
file_path_sample = r'D:\Alzheimer_s dataset\brain\TMT\TMT1_proteomics_brain.xlsx'
file_path_metadata = r'D:\Alzheimer_s dataset\brain\TMT\TMT1_clinical_brain.xlsx'
original_data = pd.read_excel(file_path_sample)
original_metadata = pd.read_excel(file_path_metadata)
data = original_data.copy()  #copying the data to maintain the originalilty of the data
metadata = original_metadata.copy()

print(data.shape)  #Dimension of the data 
print(metadata.shape)
data.head()   #visualization of the data
print(data.iloc[:, 0:2])
metadata.head()

data.rename(columns={'Unnamed: 0': 'Protein_ID'}, inplace=True)   #rename the first column to Protein_ID
data

Banner_intensity_columns = data.columns[1:] #Only the intensity columns
data_intensity = data.loc[:, Banner_intensity_columns]
data_intensity
data_protein_ID = data.loc[:, 'Protein_ID'] #only the protein ID column
data_protein_ID

df_log_transformed = np.log1p(data_intensity) #log transformation of the data
print(df_log_transformed)

from sklearn.preprocessing import StandardScaler, MinMaxScaler # Normalization of data
scaler = StandardScaler() # Standardization (Z-score normalization)
df_standardized = scaler.fit_transform(df_log_transformed)
print(df_standardized)
min_max_scaler = MinMaxScaler() # Min-Max Normalization
df_min_max_scaled = min_max_scaler.fit_transform(df_log_transformed)
print(df_min_max_scaled)
data_scaled = pd.DataFrame(data=df_min_max_scaled, columns=data_intensity.columns) #scaling
print(data_scaled) 

# Visualizing the normalized data
plt.figure(figsize=(20, 15))
sns.histplot(df_standardized.flatten(), kde=True, bins=100,) #bins100 to make the plot more clear
plt.title("Distribution of Standardized Data")
plt.show()
plt.figure(figsize=(20, 15))
sns.histplot(df_min_max_scaled.flatten(), kde=True, bins=100) 
plt.title("Distribution of Min-Max Normalized Data")
plt.show()

#Count NaN values per row
nan_counts = data_scaled.isna().sum(axis=1).sum()   
print(nan_counts) #there are totally 158 rows contating nan values in the data

# rows with nan values are removed from a dataframe
data_filtered = data_scaled.dropna(axis=0)
print(data_filtered)

from scipy.stats import zscore
z_scores = data_filtered.apply(zscore) # Z-scores
print(z_scores)
threshold = 3 # threshold for identifying outliers
rows_to_keep = (z_scores < threshold).all(axis=1) # Rows to keep
data_no_outliers = data_filtered[rows_to_keep] # Data with no outliers
data_no_outliers
print(data_no_outliers)

print(f'Original data shape: {data.shape}') #shape of the data before preprocessing
print(f'Data shape after sclaing: {data_scaled.shape}') #shape of the data after scaling
print(f'Data shape after nan values removal: {data_filtered.shape}') #shape of the data after removing nan values
print(f'Data shape after outlier removal: {data_no_outliers.shape}') #shape of the data after removing outliers

processed_df = data_no_outliers.merge(data_protein_ID, left_index=True, right_index=True) # Merging the preprocessed data with protein ID

# Reordering the columns
cols = processed_df.columns.tolist()
new_cols = [cols[-1]] + cols[:-1]
processed_df = processed_df[new_cols]
processed_df
df1_transposed = processed_df.set_index(processed_df.columns[0]).T  # Transpose the processed dataframe and set the index to the first column (Sample)
df1_transposed

# Merge the transposed dataframe with the metadata
merged_df = pd.merge(df1_transposed, metadata.set_index(metadata.columns[0]), left_index=True, right_index=True)
merged_df
```
